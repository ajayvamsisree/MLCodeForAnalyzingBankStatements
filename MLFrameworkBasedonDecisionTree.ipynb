```python
#importing required libraries and modules for using decision tree algorithm
import pandas as pd
import numpy as np
import scipy as sp
from sklearn.cross_validation import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
data_set = pd.read_excel("DA_Test.xlsx")
data_set.head()
```

```python
data_set.describe()
```

```python
#method to create dicitionary for Basis of classification
def createdict(x):
    mydict = {}
    j=0
    for i in x:
        mydict[i] = j
        j=j+1
    return mydict
```

```python
#identify the source variable debit amount,credit amount and  basis of classification
source1 = pd.DataFrame(data_set,columns=['Debit Amount','Credit Amount'])
x = source1.values
source2 = pd.DataFrame(data_set,columns = ['Basis of Classification'])
x1 = data_set['Basis of Classification'].unique()
mydict = createdict(x1)
print(mydict)
```

```python
#removing null values and replacing the basis of classification with index numbers
values = []
for i in data_set['Basis of Classification']:
    values.append(mydict[i])
source1['Basis of Classification'] = values
source1['Debit Amount'] = source1['Debit Amount'].fillna(0)
source1['Credit Amount'] = source1['Credit Amount'].fillna(0)
x = source1.values
```

```python
#storing the target variable First Level Classification 
target = pd.DataFrame(data_set,columns=['First Level Classification'])
y = target.values
```

```python
#Splitting the data in 80% train and 20% test data
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size = 0.2, random_state = 100)
```

```python
#create the decisiontree classifier based on gini index and training the algorithm with the train data
clf_gini = DecisionTreeClassifier(criterion = "gini", random_state = 100,
                               max_depth=3, min_samples_leaf=5)
clf_gini.fit(x_train, y_train)
```

```python
#predicting the test data
y_pred = clf_gini.predict(x_test)
```

```python
#print the test data and predictions
print("Debit Amount\tCredit Amount\tBasis of Classification\tFirst level classification")
for i in range(len(x_test)):
    print(x_test[i][0],"\t\t",x_test[i][1],"\t\t",list(mydict.keys())[list(mydict.values()).index(x_test[i][2])],"\t\t",y_pred[i])
#print(x_test,"\t",y_pred)
```

```python
#print accuracy score
accuracy_score(y_test,y_pred)
#for decision tree accuracy score is low so we choose another algorithm randomforest algorithm attached code in next file 
#go through it
```
